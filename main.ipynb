```python
from openai import OpenAI
import openai
from dotenv import load_dotenv
import os


load_dotenv()
openai.api_key = os.getenv("OPENAI_API_KEY")

client = OpenAI()
```

```python
response = client.completions.create(
    model = 'davinci-002',
    prompt = f"""
Given the response provided, craft three distinct paragraphs with titles for each, adhering to the following guidelines:

1.Abstract: Summarize the core essence of the response in 30-40 words, providing a brief overview.
2.Research and Discussion: In 30-40 words, elaborate on the context and delve into the key points or themes presented in the response, highlighting any relevant research or discussions.
3.Results and Conclusion: Conclude the response by presenting the outcomes or findings derived from the discussion. Offer a concise conclusion or reflection on the topic discussed within 30-40 words.

Ensure each paragraph is coherent, concise, and effectively captures the essence of its respective title.

Prompt: {}""",
    max_tokens=125,
    temperature=0.6,
    frequency_penalty=0.8
)
```
